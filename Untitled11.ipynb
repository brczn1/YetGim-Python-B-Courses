```python
#liste içerisindeki sayıları tek tek kontrol eden ve tek mi çift mi yazankodu yaz
sayılar=[15,21,78,23,44,34,41,39,55,210,147,258,369]
for sayı in sayılar :
  if sayı %2 == 0:
    print('çift sayı')
  else:
    print('tek sayı')
```

```python
#liste içerisinde 100 den büyük sayı gelirse döngüyü durdur
for sayı in sayılar:
  if sayı > 100:
    break
  print(sayı)
```

```python
#liste içindeki çift sayıları göster
#not : tek sayıları atlasın
for sayı in sayılar :
  if sayı %2 ==1:
    continue
  print(sayı)
```
