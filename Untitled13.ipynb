```python
#pandas kütüphanesi
import numpy as np
import pandas as pd
```

```python
data = {
    'öğrenci' : ['ayşe','fatma','ali','ömer','buse','evren','zeynep'],
    'matematik' :[85,65,35,25,95,90,75],
    'fizik' : [42,51,38,64,75,30,15],
    'yas ' :[17,18,17,17,17,19,19],
    'sınıf' : ['11a','11b','11c','11b','11b','11a','11c']}
```

```python
print (data)
```

```python
df = pd.DataFrame(data)
df
```

```python
#df.head(n) ilk n veriyi gösterir.
#tail(n)sondan n veriyi getirir.
df.head(3)
df.tail(2)
```

```python
df.describe() #özetler ortalamalar vs gçsteriyo.
```

```python
# sutün seçme
df['matematik']
```

```python
#birden fazla kolon seçme
df[['öğrenci','matematik','fizik']]
```

```python
#satır bazında filtreleme
#matematik puanları 80 den büyük olan bilgileri getir

df[df['matematik']>80 ]
```

```python
#matematik dersinden 70 den fazla alanlar aynı zamanda fizikten de 70 den fazla alanların listesi
# and -> &
# or -> |

df[(df['matematik']>70)&( df['fizik']>70)]
```

```python
df[df['matematik']>70 & df['fizik']>70]
```

```python
#11a sınıfındaki öğrencilerin listesi

df[df['sınıf']=='11a']
```

```python
#yeni bir kolon ekleme işlemi
df['ortalama']=( df['matematik']+df['fizik']) /2
df
```

```python
df['cinsiyet'] = ['k','k','e','e','k','e','k']
df
```

```python
#ortala sütununa göre artan bir şekilde sıralayınız
df.sort_values('ortalama')
```

```python
#azalan şekilde sıralama
df.sort_values('ortalama',ascending=False)
```

```python
#öğrenci ismine göre sıralama

df.sort_values('öğrenci',ascending=False)
```

```python
#gruplama
#mean(),count(),min(),max(),sum()
#sınıflara göre ortalama hesaplama
df.groupby('sınıf')['ortalama'].mean()
```

```python
#sınıflara göre öğrenci sayısı
df.groupby('sınıf')['öğrenci'].count()
```

```python
#cinsiyete göre kaç adet öğrenci vardır.
df.groupby('cinsiyet')['öğrenci'].count()
```

```python
#yaşa göre ortalama not ortalamasını gösteriniz.

df.groupby('Yaş')['Ortalama'].mean()
```

```python
#yaşa göre ortalama not ortalaması en büyük olan notlaro gösteriniz.

df.groupby('yaş')['ortalama'].max()
```

```python
#en başarılı öğrenciyi bulma
df.loc[df['ortalama'].idxmax()]
```

```python
#en düşük matematik notu kimde
df.loc[df['matematik'].idxmin()]
```
