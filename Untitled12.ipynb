```python
# yalın olarak python ile kodlayalım
baslangic = time.time()
A = list(range(1000001))
B = list(range(1000001))
C = []
for i in range(len(B)):
  toplam = A[i] + B[i]
  C.append(toplam)

bitis = time.time()
sure = bitis- baslangic
print('saf python ile çalışma süresi:',sure)

veriler = ['ahmet','ömer','faruk','doğan']
#             0   ,  1   ,   2   ,    3
print(veriler[0])

for veri in veriler:
  print(veri)

for i in range(len(veriler)):
  print(veriler[i])
```

```python
#numpy ile çözümü
baslangic = time.time()

A = np.arange(1000001)
B = np.arange(1000001)
C = A + B

bitis = time.time()
sure = bitis- baslangic
print('numpy ile çalışma süresi:',sure)
```
