```python
# liste fonksiyonları :
# append(veri): ilgili listeye sondan veri ekler.
teams = ['fenerbahçe','galatasaray','beşiktaş',' trabzonspor']
teams.append('orduspor')
print(teams)
```

```python
# kullanıcıdan kaç adet takım girlmesi isteniyorsa o bilgiyi al ve o takım kadar takım ekleyelim
takim_sayisi=int(input('kaç adet takım girmek istersiniz:'))
for i in range(takim_sayisi):
  takim_adi = input('Takım Adı:')
  teams.append(takim_adi)
```

```python
print(teams)
```

```python
#remove('veri): veri siler
teams.remove('fenerbahçe')
print(teams)
```

```python
#sort:artan veya a-z sıralaması yapar
teams.sort()
print(teams)
```
